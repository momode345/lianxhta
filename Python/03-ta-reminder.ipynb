````python
# 助教推文进展问询

import re
from urllib.parse import unquote, quote
import pyperclip  # 用于复制内容到剪贴板

# 输入的 URL
url = '''
OK-B808-李嘉豪：论文精要-sfma-非参数稳健SFA模型-两篇推文的工作量.md
'''.strip()

# 判断 URL 是否以 http 开头
if url.startswith('http'):
    final_url = url
else:
    # 如果没有以 'http' 开头，生成新的 URL
    final_url = "https://github.com/arlionn/lianxhta/blob/main/blog_topic/" + quote(url)

# 从 URL 中提取文件名部分
filename = final_url.split("/")[-1]

# 解码文件名中的中文字符
decoded_filename = unquote(filename)

# 提取文件名中的作者姓名（假设作者姓名是文件名中间的部分）
# 以 "：翻译：" 作为分隔符进行分割
parts = decoded_filename.split("：")
fullname = parts[0].split("-")[-1]  # 获取作者的姓名部分

# 提取姓和名
if len(fullname) == 3:  # 如果姓名是三个字，姓和名
    sname = fullname[1:]  # 取名字的部分
else:  # 否则姓就是第一个字，名是其余部分
    sname = fullname

# 提示文本
info = f'''
{sname}，你这篇推文进展如何？

{decoded_filename}

{final_url}

提交说明：参见 https://github.com/arlionn/lianxhta 第 2.3 小节。

往期推文范本：
https://github.com/arlionn/lianxhta/tree/main/template
'''

# 将提示文本复制到剪贴板
pyperclip.copy(info)

# 打印结果（供调试）
print(info)
````
